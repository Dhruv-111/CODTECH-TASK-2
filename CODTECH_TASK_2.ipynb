```python
import random
import time

class RedTeam:
    def __init__(self, target_network):
        self.target_network = target_network

    def execute_attack(self):
        # Simulate various types of attacks with varying sophistication
        attack_types = ["Phishing", "DDoS", "Brute Force", "Malware Injection"]
        sophistication_levels = ["Low", "Medium", "High"]
        attack_type = random.choice(attack_types)
        sophistication = random.choice(sophistication_levels)
        target = random.choice(self.target_network.devices)
        print(f"Red Team launches {sophistication} {attack_type} attack on {target}")

class BlueTeam:
    def __init__(self, target_network):
        self.target_network = target_network
        self.score = 0

    def monitor_network(self):
        # Monitor network traffic and devices for suspicious activity
        print("Blue Team monitors network for suspicious activity")

    def respond_to_threats(self):
        # Respond to detected threats with appropriate defense strategies
        print("Blue Team responds to detected threats")
        # Introduce a random chance for the blue team to fail in responding to threats
        if random.random() < 0.3:  # 30% chance of failure
            print("Blue Team failed to respond effectively.")
            return False
        self.score += 1  # Increment score for each detected threat
        return True

    def display_score(self):
        print(f"Blue Team Score: {self.score}")

class Network:
    def __init__(self, devices):
        self.devices = devices

# Define network devices
devices = ["Server", "Router", "Firewall", "Workstation", "Printer"]

# Create network and teams
network = Network(devices)
red_team = RedTeam(network)
blue_team = BlueTeam(network)

# Simulate cybersecurity exercise
rounds = 10
for round_num in range(1, rounds + 1):
    print(f"Round {round_num}:")
    # Red team executes attack
    red_team.execute_attack()
    # Blue team monitors network
    blue_team.monitor_network()
    # Blue team responds to threats
    if not blue_team.respond_to_threats():  # Check if the blue team response fails
        print("Red Team scores!")
    # Display blue team score after each round
    blue_team.display_score()
    print("-" * 50)
    # Wait for a random time interval before the next iteration
    time.sleep(random.uniform(1, 3))

print("Cybersecurity exercise completed.")
```
